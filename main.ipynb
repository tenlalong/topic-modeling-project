```python
# net.vonforst.evmap

df11 = pd.read_csv("net.vonforst.evmap.csv")

feature1 = df11['Review']

# Remove punctuations

import string
text_features=''

for i in range(0,len(feature1)):
    text_features += "".join([char for char in feature1[i] if char not in string.punctuation]) + ' '

    
    
# Remove special characters and emojis

from nltk import word_tokenize

wordFeature_token = word_tokenize(text_features)

output = ''

for i in range(0,len(wordFeature_token)):
    if wordFeature_token[i].isalnum():
        output+=wordFeature_token[i] + ' '
        
        
# Convert numbers into text - please run "pip install num2words" in your terminal before running this

from num2words import num2words

for i in range(0, len(wordFeature_token)):
    if wordFeature_token[i].isnumeric():
        wordFeature_token[i] = num2words(wordFeature_token[i])
        
output = ''

for i in range(0,len(wordFeature_token)):
    if wordFeature_token[i].isalnum():
        output+=wordFeature_token[i] + ' '
        

        # Remove extra white spaces, show the output
for i in range(0, len(wordFeature_token)):
    if wordFeature_token[i].isnumeric():
        wordFeature_token[i] = num2words(wordFeature_token[i])
        
output = ''

for i in range(0,len(wordFeature_token)):
    if wordFeature_token[i].isalnum():
        output+=wordFeature_token[i]

        
# Turns all words to lowercase, show the output
output = output.lower()


# Stopword filtering

for i in range(0, len(wordFeature_token)):
    if wordFeature_token[i].isnumeric():
        wordFeature_token[i] = num2words(wordFeature_token[i])

for i in range(0,len(wordFeature_token)):
    wordFeature_token[i] = wordFeature_token[i].lower()
    
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
filtered_words = [word for word in wordFeature_token if word not in stop_words]


# Lemmatization
lemma = nltk.stem.WordNetLemmatizer()
lemmatized1 = [lemma.lemmatize(word) for word in filtered_words]

print(lemmatized1)
```

### Bonus of Task 2

1) The fact that the list of stop words in NLTK can be customized does NOT affect the analysis of the reviews. Stop words are defined as repeating words that do not contribute to the importance of the sentence/document/text/paragraph. 

==> The help of NLTK stop words only filters out the unnecessary words from the reviews, and this would help us with Sentiment Analysis later on. 

2) If we use a list of non-customized list of stop words, some (or all) of the extra words that does not contribute to the review won't get filtered, which may affect the Sentiment Analysis process.

### Comparison between TextBlob and Vader

Textblob is a simple Python library offering API access to sentiment analysis. Textblob sentiment analyzer returns the polarity, which is a float that lies between \[-1 and 1\]. While -1 indicates negative sentiment, +1 indicates positive sentiments.

Valence aware dictionary for sentiment reasoning (VADER) uses a list of words which are labelled positive or negative according to their semantic orientation to calculate the next sentiment. Vader semtiment returns the probability of a given input sentence to be either positive, negative, or neutral. The sum of the 3 probabilities will be 100%.

For this project, I feel Vader is the better option, because we can analyze in detail on how positive/negative/neutral a review would be, and thus knowing more about customer satisfaction on an application. In addition, if we do not use NLTK stop words, TextBlob will not give accurate score, while Vader is also a customized set of words (similar to NLTK).

```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]
data = papers.Review_processed.values.tolist()
data_words = list(sent_to_words(data))
# remove stop words
data_words = remove_stopwords(data_words)
print(data_words[:1][0][:30])
```

```python
from pprint import pprint
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
