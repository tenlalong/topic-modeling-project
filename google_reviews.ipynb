```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
# Scrape the reviews for com.fortum.evplus

from google_play_scraper import Sort, reviews_all

us_reviews = reviews_all(
    'com.fortum.evplus',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

```python
df_busu = pd.DataFrame(np.array(us_reviews),columns=['review'])


df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))

```

```python
df = pd.DataFrame({'Package name': 'com.fortum.evplus',
                   'Reviewer name': df_busu.userName,
                   'Review': df_busu.content,
                   'Rating': df_busu.score
    
})
df.to_csv('com.fortum.evplus.csv')
```
